```python
#------------ USANDO CNN to classify CIFAR images. -----------------
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

'''
The CIFAR10 dataset contains 60,000 color images in 10 classes, 
with 6,000 images in each class. The dataset is divided into 
50,000 training images and 10,000 testing images. 
The classes are mutually exclusive and there is no overlap
between them.
'''
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
#Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

#Verify the dataset images. Plottingg 25 images
class_names = ['airplane', 'automobile' , 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'shihp', 'truck']


plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()

#CRIANDO A BASE CONVULACIONAL
#A entrada da CNN serah 32 x 32 x 3, onde img_height x img_width x color_channels
#Layers Convolutional and Pooling
model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3), activation='relu'))

#Layer Fully complete
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

#Compilando e treinando o modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10,
                   validation_data=(test_images, test_labels))

model.summary()

#Plotting of evaluation
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```

```python
#USANDO UMA REDE PARA CLASSICATION

from __future__ import absolute_import, division, print_function, unicode_literals

#TensorFlow e tf.keras
import tensorflow as tf
from tensorflow import keras

#Libraries Auxiliares
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)

#Importando Fashion MNIST diretamente do Tensorflow
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

class_names = ['T-Shirt/Top', 'Trouser', 'Pullover', 'Dress', 'Coat',
                'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

#Mostrando qtd de imagens e seu tamanho no treinamento
train_images.shape

#Mosrando labels no conjunto de trainamento
len(train_labels)

train_labels

#Mostrando qtd de imagens do Test
test_images.shape

#Mosrando labels no conjunto de Teste
len(test_labels)

test_labels

#Plotando imagem antes de pré-processados. Note que esta de 0 a 255
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()

#Escalaremos esses valores entre 0 e 1
train_images = train_images / 255.0
test_images = test_images / 255.0

#Antes de construir a rede, vejamos as imagens
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()

#Construindo o modelo
#Montando as camadas
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #1º camada da rede, transforma array de 2D para 1D=784pixels
    keras.layers.Dense(128, activation='relu'), #2º camada com 128 nós
    keras.layers.Dense(10, activation='softmax') #3º camda com 10 nós
])

#Compilando o modelo
model.compile(optimizer='adam', #como o modelo se atualiza com base no dado que ele ve e sua funcao loss
              loss='sparse_categorical_crossentropy', #Esse mede quão precisa o modelo é durante o treinamento.
              metrics=['accuracy']) #para monitorar os passos de treinamento e teste

#Treinando o modelo
#Para começar a treinar use o método "model.fit". Este vai encaixar o modelo no conjunto de treinamento
model.fit(train_images, train_labels, epochs=10)

#Avaliando acurácia
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nTest Accuracy:', test_acc)

#Fazendo predições
predictions = model.predict(test_images)

#A predição eh um array de 10 numeros, que representa a confiança
#do modelo que a imagem corresponde a cada um dos artigos de roupa
predictions[0] #verifiacando o item 0

np.argmax(predictions[0]) #mostrando o maior indicie de confianca
test_labels[0] #O numero de predicao eh igual ao label
```
